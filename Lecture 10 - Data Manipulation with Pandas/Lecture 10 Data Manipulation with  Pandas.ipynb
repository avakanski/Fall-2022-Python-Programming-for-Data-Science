# 10.1 Introduction to pandas <a id="section1"/>

`pandas` is a library designed for working with structured data, such as tabular data (e.g., from .csv files, Excel files) or relational databases (e.g., SQL). 

> The ***DataFrame*** object in `pandas` is a 2-dimensional tabular, column-oriented data structure. The DateFrame is similar to a spreadsheet and can store data of different types (including text characters, integers, floating-point values, categorical data, and more).

<img style="float: center; height:260px;" src="images/pic1.jpg">

The `pandas` name is derived from the term *panel data*, which is a term from economics for multi-dimensional structured data. 

For more information about `pandas` you can consult the full [pandas documentation](http://pandas.pydata.org/pandas-docs/stable/).

In this lecture, we will first work with example data and go through the steps of preparing data for analysis. Using `pandas` for these tasks is much more convenient than using Python's built-in data structures.

The first column with the indices in `pandas` DataFrames do not need to be a sequence of integers, but they can also be strings or other numeric data (e.g., dates, years).

For instance, let's create a `DataFrame` called *bacteria* to see how indexing with string indices works. We again pass in a dictionary, with the keys corresponding to column names and the values to the data, and in addition we pass a list of strings called `index`. In `pandas`, two other operators are used for selecting rows and/or columns: `.loc` and `.iloc`. 

`.loc` works with string labels, and `.iloc` works with integer labels. These two operators can accept a single label, a list of labels, or a slice of labels (e.g., `'a:f'` or `2:5`).

If we're interested in the row *Actinobacteria*, we can use `.loc` and the index name. This returns the column values for the specified row.

````python
bacteria.loc['Actinobacteria']
````

We could have also used "positional indexing," even though the indices are strings. The difference is that `.loc` returns a `Series` because we selected a single label, while `[2:3]` returns a `DataFrame` because we selected a range of positions.

````python
bacteria[2:3]
````

Let's return to the *unemployment* data to show how `.iloc` is used, since *unemployment* has integer indices. To select specific rows, we can do the following.

````python
unemployment.iloc[0:4]
````

````python
unemployment.iloc[[1, 5, 6, 22]]
````

We can also select a range of rows and specify the step value.

### Selecting Multiple Rows and Columns

Both `.loc` and `.iloc` can be used to select subsets of rows and columns at the same time if they are given lists (and/or slices for `.iloc`) as their two arguments. 

Using `.iloc` on the *unemployment* DataFrame, select the rows starting at row 2 and ending at row 5, and the 0th, 2nd, and 3rd columns.

````python
unemployment.iloc[2:6,[0,2,3]]
````

````python
# The same as above
unemployment.loc[2:6,['country', 'year_month', 'unemployment']]
````

# 10.4 Creating New Columns, Reordering <a id="section4"/>

Since the `year_month` column is not correct, let's try to split it into two separate columns for years and months. 

In the previous section we saw that the data type in this column is `float64`. We'll first extract the year using the `.astype()` method. This allows for type casting, i.e., using `.astype(int)` we will convert the floating point values into integer numbers (by truncating the decimals).

The new column *year* will be added on the right of the DataFrame.

````python
unemployment['year'] = unemployment['year_month'].astype(int)
unemployment.head()
````

Next, let's create a new column *month*. We will subtract the *year* value from *year_month* to get the decimal portion of the value, and multiply the result by 100 and convert to `int`. Because of the truncating that occurs when casting to `int`, we first need to round the values to the nearest whole number using `round()`. 

````python
unemployment['month'] = ((unemployment['year_month'] - unemployment['year']) * 100).round(0).astype(int)
unemployment.head(12)
````

Now, let's try to reorder the newly created *year* and *month* columns in the DataFrame. For this, we use the square brackets notation again, passing in a list of column names in the order we would like to see them.

````python
unemployment = unemployment[['country', 'seasonality',
                             'year_month', 'year', 'month',
                             'unemployment', 'unemployment_rate']]
unemployment.head(10)
````

# 10.5 Merging <a id="section5"/>

If we examine the `unemployment` DataFrame we can notice that we don't exactly know what the values in the `country` column refer to. We can fix that by getting the country names from the `countries` DataFrame that we imported earlier. 

We can see in the `countries` data that *at* stands for Austria. The data even provides the country names in three different languages.

````python
countries.head()
````

Because the data we need is stored in two separate files, we will first merge the two DataFrames. The `country` column is shown in both DataFrames, so it is a good option for joining the data. However, we don't need all of the columns in the `countries` DataFrame, and therefore, we will create a new DataFrame. To select certain columns to retain, we can use the bracket notation that we used earlier to reorder the columns.

````python
country_names = countries[['country', 'country_group', 'name_en']]
````

````python
country_names.head(5)
````

`pandas` include an easy-to-use `merge` function, which has the following syntax.

```
pd.merge(first_file, second_file, on=['column_name'])
```


````python
unemployment = pd.merge(unemployment, country_names, on=['country'])
unemployment.head()
````

If we want to merge two files using multiple columns that exist in both files, we can pass a list of column names to the `on` parameter.

For more information on merging, check the pandas [documentation](http://pandas.pydata.org/pandas-docs/stable/merging.html#database-style-dataframe-joining-merging).

# 10.6 Calculating Unique and Missing Values <a id="section6"/>

In the *unemployment* data, we might want to know what countries we have data for. To extract this information, we can use the `.unique()` method. Note that the countries are listed in the right-most column `name-en` so we will use it to find the unique elements in that column.

````python
unemployment.head()
````

````python
unemployment.name_en.unique()
````

To get a count of the **number of unique countries,** we can use the `.nunique()` method.

````python
unemployment.name_en.nunique()
````

Or, we can also use `len()` to get the number of items in the array

````python
len(unemployment.name_en.unique())
````

It might be interesting to know **how many observations** we have per country. `pandas` have a method called `.value_counts()` that returns the counts for the unique values in a column.

````python
unemployment['name_en'].value_counts()
````

By default, the output is sorted by values. If we would like it sorted by index (or, country name in this case), append the `.sort_index()` method.

````python
unemployment['name_en'].value_counts().sort_index()
````

As we noticed earlier, there are missing values in the `unemployment_rate` column. To find out **how many unemployment rate values are missing** we will use the `.isnull()` method, which returns a corresponding boolean value for each missing entry in the unemployment_rate column. As we know, in Python `True` is equivalent to 1 and `False` is equivalent to 0. Thus, when we add the result with `.sum()`, we get a count for the total number of missing values.

````python
unemployment['unemployment_rate'].isnull().sum()
````

### GroupBy

What if we would like to know how many missing values exist at the *country* level? We can take the main part of what we had above and create a new column in the `DataFrame`. This is the last column, in which `False` means that the value is not missing.

````python
unemployment['unemployment_rate_null'] = unemployment['unemployment_rate'].isnull()
unemployment.head()
````

To count the **number of missing values for each country,** we can use the `.groupby()` method. It groups the data by the country `name_en` column included in the parentheses, and `.sum()` operation is performed over the `unemployment_rate_null` column.

````python
unemployment.groupby('name_en')['unemployment_rate_null'].sum()
````

# 10.8 Dealing With Missing Values: Boolean Indexing <a id="section8"/>

There are two main options for dealing with missing values:

* Fill the missing values with some other values.
* Do not use observations with missing values.
    
Here we will adopt the second approach and **exclude missing values** from our primary analyses. Instead of just getting rid of that data, it might make sense to store it in a separate DataFrame. This way, we could answer questions such as, "do missing values occur during certain months (or years) more frequently?" 

To do this, we will use *boolean indexing* for filtering data.

Recall that `unemployment['unemployment_rate'].isnull()` produces an array of boolean values. We used this previously when counting the number of missing values. This is shown in the next cell.

````python
unemployment['unemployment_rate'].isnull()[:10]
````

To create `unemployment_rate_missing`, we will index `unemployment` with the array above. This returns only the rows where the value in the array is `True`.

````python
unemployment_rate_missing = unemployment[unemployment['unemployment_rate'].isnull()]
unemployment_rate_missing.head()
````

It is also possible to specify multiple conditions using the `&` operator, but each condition needs to be inside of parentheses. 

Now, to remove the missing data in `unemployment`, we can use the `.dropna()` method. This method drops all observations for which `unemployment_rate == NaN`.

````python
unemployment.dropna(subset=['unemployment_rate'], inplace=True)
unemployment.head()
````

### Sorting Values

````python
unemployment.sort_values('unemployment_rate', ascending=False)[:10]
````
